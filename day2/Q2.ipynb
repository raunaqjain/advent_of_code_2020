```python
from collections import Counter
```

```python

with open('Q2.txt') as f:
    data = f.read().split("\n")
```

```python
ans = 0
for row in data:
    limit, char, string = row.split()
    count = Counter(string)
    if char[0] in count:
        a,b = limit.split('-')
        if int(a) <= count[char[0]] <= int(b):
            ans += 1
print (ans)
```
